```python
# Import dependencies
import pandas as pd
```

```python
# Converting csv to pandas dataframe
medal_counts_df = pd.read_csv("modified_data\Medal Counts Country Codes.csv", header=0)
medal_counts_df.head()
```

```python
# Converting csv to pandas dataframe
country_lat_long_df = pd.read_csv("original_data\countries.csv", header=0)
country_lat_long_df.info()
```

```python
# Removing unneeded columns and naming columns to be merged
country_lat_long_df_cleaned = country_lat_long_df.drop(country_lat_long_df.columns[[0,1,2,6]],axis=1)
country_lat_long_df_cleaned.columns = ['Country Code','latitude','longitude']
country_lat_long_df_cleaned['lat_long'] = country_lat_long_df_cleaned['latitude'].astype(str) + ', ' + country_lat_long_df_cleaned['longitude'].astype(str)
country_lat_long_df_cleaned.head()
```

```python
# Merging dataframes
merged_df = pd.merge(country_lat_long_df_cleaned,medal_counts_df,on='Country Code')
display(merged_df)
```

```python
# Creating JSON file of the dataframe
json_data = merged_df.to_json(orient='records')
with open('modified_data/merged_medal_counts_latlong.json', 'w') as json_file:
    json_file.write(json_data)
```
